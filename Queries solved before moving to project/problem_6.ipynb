```python
df = pd.read_csv(r'D:\Python\Python course\data_jobs.csv')
dfvg = pd.read_csv(r'D:\Python\Alex\Files\vgsales.csv')

new_columns = ['job', 'fname', 'location', 'via', 'schedule', 'remote','search_location','posted_dt','no_degree','health_ins','country','salrate','salyr','salhr','company','skills','skilltype']
df.columns = new_columns

# data clean up
df['posted_dt'] = pd.to_datetime(df['posted_dt'])                                                                # converting the column from string format to date time format
df['skills'] = df['skills'].apply(lambda gg: ast.literal_eval(gg) if pd.notna(gg) else gg)                       # converts job_skills from sting to list
#df.dropna(subset=['salyr'],inplace=True)
```

##### prob 6

###### <pre>get pie charts for work from home, no degree mention, job health insurance. in a single fig

###### solution

```python
df_6 = df.copy()
```

```python
df_6.sample(2)

jkg = ['remote','no_degree','health_ins']
dict = {gg : df_6[gg].value_counts() for gg in jkg}
```

```python
fig,ax = plt.subplots(1,len(jkg))
fig.set_size_inches(9,5)

for i,list in enumerate(jkg):
    dict[list].plot(kind='pie', y= 'count', ax = ax[i], title = list, startangle= 90, autopct='%1.1f%%' ) 
    ax[i].set_ylabel('')


plt.tight_layout()
```

```python

benefits = {
    'remote': 'Work from Home Offered',
    'no_degree': 'Degree Requirement',
    'health_ins': 'Health Insurance Offered'
}


dict = {gg : df_6[gg].value_counts() for gg in benefits.keys()}

fig,ax = plt.subplots(1,len(benefits))
fig.set_size_inches(9,5)

for i,(key,value) in enumerate(benefits.items()):
    dict[key].plot(kind='pie', y= 'count', ax = ax[i], title = value, startangle= 90, autopct='%1.1f%%' )
    ax[i].set_ylabel('')


plt.tight_layout()
```
